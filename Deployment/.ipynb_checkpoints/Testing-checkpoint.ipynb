```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os
import time
```

```python
current_dir = os.getcwd()
```

```python
os.chdir("/tf/FoodClassifier/Training")
model = tf.keras.models.load_model('saved_model/my_model')
```

```python
def preprocess(file_name):
    img_raw=tf.image.decode_image(open(file_name, 'rb').read(), channels=3)
    img = tf.expand_dims(img_raw, 0)
    img_input = tf.image.resize(img, (224, 224))/255
    
    return img_input
```

```python
file = tf.keras.utils.get_file(
    "waffles.jpg",
    "https://www.gastronomistas.com/wp-content/uploads/Destacada-La-real-1.jpg")
```

```python
os.chdir(current_dir)
img = preprocess(file)
```

```python
plt.imshow(img[0,:,:,:])
```

```python
result = model(tf.constant(img))
```

```python
tf.argmax(result,axis=1)
```
